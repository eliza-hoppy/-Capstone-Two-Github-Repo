```python
#import all packages needed
import pandas as pd
import json
```

```python
#Explore contents of 'food-inspection-violations_1.json'

#Load JSON file restaurant-and-services_2.json with a context manager in read mode
with open('food-inspection-violations_1.json', 'r') as file:
    json_1 = json.load(file)

#Check data type of the json_2 file
print(type(json_1))

#Inspect structure of json_1
json_1
```

```python
# Extract the keys from first element of json_1 to create column headings for DataFrame.

# Create an empty list for the DataFrame
col_names = []

# Extract first dictionary for the list, json_1
dict_1 = json_1[0]

# Extract each key from dict_1 and store it in list 'col_names'

for key in dict_1.keys():
    col_names.append(key)
    df_col = pd.DataFrame(col_names)
print(col_names)
print(df_col)
```

```python
# Extract values of each dictionary in json_1 and create a DataFrame

# Create empty DataFrame for all information
df1 = pd.DataFrame()
# Create empty list for values of dictionary
vals = []

for dictionary in json_1:
    for val in dictionary.values():
        vals.append(val)
    df_vals = pd.DataFrame(vals)
    
    # Attach new values to df1
    pd.concat([df1,df_vals])

print(df1.head())
```

# Data Organization

```python
#Isolate the features from json_4
json_4_ft=json_4['features']

#Print out json_4
json_4_ft
```

```python
#Isolate the first element from json_4
elm_1 = json_4_ft[0]


#Print the type of the first element
print(type(elm_1))

#Print the first element
print(elm_1)

#Print the keys from the first element
print(elm_1.keys())
```

```python
#Create an empty list for the keys in the first element
df4_col = []

#Collect the nested keys within geometry and add them to df4_col
for keys in elm_1['geometry'].keys():
    df4_col.append(keys)

#Collect the nested keys within properties and add them to df4_col
for keys in elm_1['properties'].keys():
    df4_col.append(keys)

#Print the full list of nested keys
print(df4_col)

#Find the number of keys in df4_col and df3_col
print(len(df4_col))
print(len(df3_col))
```

```python
#Create a set of column names in df3
s = set(df3_col)

#Find out what elements in df4_col are different from df3 column. 
#This will be used to determine if the two df3 and df4 are different.
diff = [x for x in df4_col if x not in s]
print(diff)
```

```python
#Remove the differing keys from key_list and df3_col.
df4_col.remove('coordinates')
df4_col.remove('type')
df3_col.remove('geolocation')
df4_col.append('geolocation.lat')
df4_col.append('geolocation.lon')

#Compare the contents of key_list to df3_col
a = sorted(df4_col)
b = sorted(df3_col)
c = zip(a,b)
print(list(c))
```

```python
#Build DataFrame df_4 from contents of json_4_ft

#Create an empty DataFrame df_4
df_4 = pd.DataFrame()

#Create empty list coord_list
coord_list = []

#Loop through each entry in json_4_ft
for entry in json_4_ft:
    #Append coordinated from the nested dictionary to empty list coord_list
    coord_list.append(entry['geometry']['coordinates'])
    
    #Turn coord_list to a DataFrame
    df_coord = pd.DataFrame(coord_list)
    
    #Append df_coord as row the df_4
    pd.concat([df_4,df_coord])
print(df_4)
```
